```python
aprobados = df_alumnos['Nota']>=7
aprobados
```

```python
df_alumnos_aprobados = df_alumnos[aprobados]
df_alumnos_aprobados
```

```python
df_alumnos_aprobados.to_csv('alumnos_aprobados.csv', index=False)
```

```python
alumnos['Puntos_extras'] = alumnos['Nota'].apply(lambda x: x*0.4)
alumnos
```

```python
alumnos['Notas_finales'] = alumnos['Nota'] + alumnos['Puntos_extras']
alumnos
```

```python
alumnos['Aprobado_final'] = alumnos['Notas_finales'].apply(lambda x: True if x >= 7.0 else False)
alumnos
```

```python
alumnos.query('Aprobado == False & Aprobado_final == True')
alumnos
```

```python
seleccion = (alumnos['Aprobado'] == False) & (alumnos['Aprobado_final'] == True)
alumnos[seleccion]
```
