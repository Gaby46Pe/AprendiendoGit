# 4. Productos más y menos vendidos

```python
productos_t1 = display(tienda.groupby('Producto').size().sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

productos_vendidos = tienda.groupby('Producto').size().sort_values(ascending=False)
mas_vendidos = productos_vendidos.max()
menos_vendidos = productos_vendidos.min()
productos_mas_vendidos = productos_vendidos[productos_vendidos == mas_vendidos]
productos_menos_vendidos = productos_vendidos[productos_vendidos == menos_vendidos]

print("Productos más vendidos en Tienda 1:")
display(productos_mas_vendidos)

print("\nProductos menos vendidos en Tienda 1:")
display(productos_menos_vendidos)

productos_grafico = pd.concat([productos_mas_vendidos, productos_menos_vendidos])

colors = ['purple'] * len(productos_mas_vendidos) + ['yellow'] * len(productos_menos_vendidos)

plt.figure(figsize=(8, 4))
productos_grafico.plot(kind='barh', color=colors)
plt.title('Productos Más y Menos Vendidos en Tienda 1')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.tight_layout()
display(plt.show())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
# Cargar datos de las 4 tiendas

# Crear un diccionario para facilitar el análisis
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
```

```python
ingresos = {}
for nombre, df in tiendas.items():
    ingresos[nombre] = df['Precio'].sum()

# Mostrar los ingresos
print(ingresos)
```

```python
# Datos para gráfico
tiendas_names = list(ingresos.keys())
ingresos_values = list(ingresos.values())

plt.figure(figsize=(8,6))
plt.bar(tiendas_names, ingresos_values, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos ($)')
plt.show()
```

```python
ventas_por_categoria = {}
for nombre, df in tiendas.items():
    ventas_categoria = df.groupby('Categoría del Producto').size().sort_values(ascending=False)
    ventas_por_categoria[nombre] = ventas_categoria

# Ejemplo para visualizar una tienda
for nombre, ventas in ventas_por_categoria.items():
    print(f"\nVentas por categoría en {nombre}:\n{ventas}")
```

```python
# Ejemplo para una tienda
nombre_tienda = 'Tienda 1'
ventas_categoria = ventas_por_categoria[nombre_tienda]
top_categorias = ventas_categoria.head(5)

plt.figure(figsize=(8,8))
plt.pie(top_categorias, labels=top_categorias.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Principales Categorías en {nombre_tienda}')
plt.show()
```

```python
calificaciones = {}
for nombre, df in tiendas.items():
    calificaciones[nombre] = df['Calificación'].mean()

print("Calificación promedio por tienda:")
for nombre, cal in calificaciones.items():
    print(f"{nombre}: {cal:.2f}")
```

```python
calificaciones_list = list(calificaciones.values())

plt.figure(figsize=(8,6))
plt.bar(tiendas_names, calificaciones_list, color='orange')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5)  # Asumiendo escala de 1 a 5
plt.show()
```

```python
productos_vendidos = {}
for nombre, df in tiendas.items():
    conteo_productos = df['Producto'].value_counts()
    productos_vendidos[nombre] = {
        'más_vendidos': conteo_productos.head(5),
        'menos_vendidos': conteo_productos.tail(5)
    }

# Ejemplo de visualización
for nombre, productos in productos_vendidos.items():
    print(f"\n{nombre} - Productos más vendidos:\n{productos['más_vendidos']}")
    print(f"\n{nombre} - Productos menos vendidos:\n{productos['menos_vendidos']}")
```

```python
envios_promedio = {}
for nombre, df in tiendas.items():
    envios_promedio[nombre] = df['Costo de envío'].mean()

print("Costo de envío promedio por tienda:")
for nombre, costo in envios_promedio.items():
    print(f"{nombre}: {costo:.2f}")
```
