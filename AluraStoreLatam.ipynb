```python
import matplotlib.pyplot as plt
import pandas as pd

productos_vendidos = tienda.groupby('Producto').size().sort_values(ascending=False)
mas_vendidos = productos_vendidos.max()
menos_vendidos = productos_vendidos.min()
productos_mas_vendidos = productos_vendidos[productos_vendidos == mas_vendidos]
productos_menos_vendidos = productos_vendidos[productos_vendidos == menos_vendidos]

print("Productos más vendidos en Tienda 1:")
display(productos_mas_vendidos)

print("\nProductos menos vendidos en Tienda 1:")
display(productos_menos_vendidos)

productos_grafico = pd.concat([productos_mas_vendidos, productos_menos_vendidos])

colors = ['purple'] * len(productos_mas_vendidos) + ['yellow'] * len(productos_menos_vendidos)

plt.figure(figsize=(8, 4))
productos_grafico.plot(kind='barh', color=colors)
plt.title('Productos Más y Menos Vendidos en Tienda 1')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.tight_layout()
display(plt.show())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
# Cargar datos de las 4 tiendas

# Crear un diccionario para facilitar el análisis
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
```

```python
ingresos = {}
for nombre, df in tiendas.items():
    ingresos[nombre] = df['Precio'].sum()

# Mostrar los ingresos
print(ingresos)
```

```python
# Datos para gráfico
tiendas_names = list(ingresos.keys())
ingresos_values = list(ingresos.values())

plt.figure(figsize=(8,6))
plt.bar(tiendas_names, ingresos_values, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos ($)')
plt.show()
```

```python
ventas_por_categoria = {}
for nombre, df in tiendas.items():
    ventas_categoria = df.groupby('Categoría del Producto').size().sort_values(ascending=False)
    ventas_por_categoria[nombre] = ventas_categoria

# Ejemplo para visualizar una tienda
for nombre, ventas in ventas_por_categoria.items():
    print(f"\nVentas por categoría en {nombre}:\n{ventas}")
```

```python
# Ejemplo para una tienda
nombre_tienda = 'Tienda 1'
ventas_categoria = ventas_por_categoria[nombre_tienda]
top_categorias = ventas_categoria.head(5)

plt.figure(figsize=(8,8))
plt.pie(top_categorias, labels=top_categorias.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Principales Categorías en {nombre_tienda}')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Lista de tiendas a comparar
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Número de categorías principales a mostrar
top_n = 5

# Preparar datos: obtener las top N categorías para cada tienda
top_categorias_por_tienda = {}
for tienda in tiendas:
    ventas_categoria = ventas_por_categoria[tienda]
    top_categorias = ventas_categoria.head(top_n)
    top_categorias_por_tienda[tienda] = top_categorias

# Construir un DataFrame para facilitar la comparación
import pandas as pd

df = pd.DataFrame()

for tienda in tiendas:
    categorias = top_categorias_por_tienda[tienda]
    df[tienda] = categorias

# Reemplazar NaNs por ceros (en caso de que alguna categoría no esté en todas las tiendas)
df = df.fillna(0)

# Crear un gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(12, 6))

# Número de categorías principales
indices = np.arange(top_n)

# Ancho de las barras
width = 0.2

# Plotear las barras para cada tienda
for i, tienda in enumerate(tiendas):
    ax.bar(indices + i * width, df[tienda], width=width, label=tienda)

# Configurar etiquetas y leyenda
ax.set_xticks(indices + width * (len(tiendas) - 1) / 2)
ax.set_xticklabels(df.index)
ax.set_xlabel('Categorías')
ax.set_ylabel('Ventas')
ax.set_title('Comparativa de Principales Categorías por Tienda')
ax.legend()

plt.tight_layout()
plt.show()
```
